# Load a datasets

```python
path1 = "data/co-presence/InVS13.dat"
path2 = "data/co-presence/InVS15.dat"
path3 = "data/co-presence/LH10.dat"

data1 = cs.load_data(path1)
data2 = cs.load_data(path2)
data3 = cs.load_data(path3)
data = {"InVS13" : data1,
        "InVS15" : data2,
        "LH10"   : data3}
```

### Spektral gap

```python
for k in list(data.keys()):
    graph = cs.build_weighted_graph(data[k],19)
    sp_bin = ut.spectral_gap(graph,
                    binary_adj_matrix=True)
    sp_wei = ut.spectral_gap(graph,
                    binary_adj_matrix=False)
    
    print("NAME: \t",k)
    print("Spectral GAP binary matrix\t",sp_bin)
    print("Spectral GAP weight matrix\t",sp_wei)
    print("\n")
```
